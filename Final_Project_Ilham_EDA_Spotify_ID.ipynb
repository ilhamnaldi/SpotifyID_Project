```python
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'])

latest_date = df['snapshot_date'].max()
date_threshold = latest_date - pd.Timedelta(days=50)

recent_df = df[df['snapshot_date'] >= date_threshold]

top_1_recent = recent_df[recent_df['daily_rank'] == 1]

top_1_counts_recent = top_1_recent[['name', 'artists']].value_counts().reset_index(name='count')

print("Top 1 songs in the last 50 days:")
top_1_counts_recent
```

```python
selected_songs = ['You\'ll Be in My Heart - Spotify Singles', 'Mangu', 'Nina', 'Aku Dah Lupa']

# Filter DataFrame for the selected songs
filtered_df = df[df['name'].isin(selected_songs)]

# Select relevant columns and drop duplicates to get unique song entries
unique_song_features = filtered_df[['name', 'artists', 'speechiness', 'acousticness', 'loudness', 'tempo', 'valence']].drop_duplicates(subset=['name', 'artists'])

print("Analisis fitur audio untuk lagu-lagu terpilih:")
unique_song_features
```

```python
# Combine acousticness and valence for each song
df['acousticness_valence'] = df['acousticness'] * df['valence']

# Get unique songs with their combined feature
unique_songs_features = df.drop_duplicates(subset=['name', 'artists'])[['name', 'artists', 'acousticness_valence', 'tempo']]

# Sort by the combined feature in descending order
top_combined_songs = unique_songs_features.sort_values(by='acousticness_valence', ascending=False)

print("Top songs by combining acousticness and valence (unique names):")
top_combined_songs
```

```python
# Combine tempo with other features for each song
df['tempo_rank'] = df['tempo']

# Get unique songs with their tempo
unique_songs_tempo = df.drop_duplicates(subset=['name', 'artists'])[['name', 'artists', 'tempo_rank']]

# Sort by tempo in descending order
top_tempo_songs = unique_songs_tempo.sort_values(by='tempo_rank', ascending=False)

print("Top songs by tempo (unique names):")
top_tempo_songs.head(16)
```

```python
# Combine energy with other features for each song
df['energy_rank'] = df['energy']

# Get unique songs with their energy
unique_songs_energy = df.drop_duplicates(subset=['name', 'artists'])[['name', 'artists', 'energy_rank']]

# Sort by energy in descending order
top_energy_songs = unique_songs_energy.sort_values(by='energy_rank', ascending=False)

print("Top songs by energy (unique names):")
top_energy_songs.head(50)
```

```python
# Get unique songs with their acousticness
# Include 'acousticness' in the columns when creating unique_songs_features
unique_songs_features = df.drop_duplicates(subset=['name', 'artists'])[['name', 'artists', 'acousticness_valence', 'tempo', 'acousticness']]

# Sort by acousticness in descending order
top_acousticness_songs = unique_songs_features.sort_values(by='acousticness', ascending=False)

print("Top 50 songs by acousticness (unique names):")
top_acousticness_songs.head(50)
```

```python
# Get unique songs with their valence
unique_songs_valence = df.drop_duplicates(subset=['name', 'artists'])[['name', 'artists', 'valence']]

# Sort by valence in descending order
top_valence_songs = unique_songs_valence.sort_values(by='valence', ascending=False)

print("Top 50 songs by valence (unique names):")
top_valence_songs.head(50)
```

```python
import seaborn as sns
# Select the audio features for boxplot
audio_features = ['speechiness', 'acousticness', 'loudness', 'tempo', 'valence']
# Create subplots for each feature
fig, axes = plt.subplots(nrows=len(audio_features), ncols=1, figsize=(10, 15))
# Create a boxplot for each audio feature
for i, feature in enumerate(audio_features):
  sns.boxplot(x=df[feature], ax=axes[i], orient='h')
  axes[i].set_title(f'Boxplot for {feature}')
  axes[i].set_xlabel(feature)
# Adjust layout to prevent overlapping titles
plt.tight_layout()
# Show the plots
plt.show()
```
